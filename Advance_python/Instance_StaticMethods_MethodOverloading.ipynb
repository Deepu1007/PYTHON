```python
class Book:
    total_books = 0  # Class variable to keep track of total books

    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        Book.total_books += 1  # Increment total books when a new book is added

    def update_title(self, new_title):
        self.title = new_title

    def update_author(self, new_author):
        self.author = new_author

    def display_info(self, user_type):
        if user_type == "librarian":
            return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}"
        elif user_type == "reader":
            return f"Title: {self.title}, Author: {self.author}"
        else:
            return "Invalid user type"

    @staticmethod
    def book_info():
        return "Books are a collection of written works."

    @classmethod
    def get_total_books(cls):
        return cls.total_books


class Author:
    total_authors = 0  # Class variable to keep track of total authors

    def __init__(self, name, birthdate):
        self.name = name
        self.birthdate = birthdate
        self.books = []
        Author.total_authors += 1  # Increment total authors when a new author is added

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, isbn):
        self.books = [book for book in self.books if book.isbn != isbn]

    @staticmethod
    def author_info():
        return "Authors are individuals who write books."

    @classmethod
    def get_total_authors(cls):
        return cls.total_authors


class Library:
    library_count = 0  # Class variable to keep track of the number of libraries

    def __init__(self):
        self.books = []
        self.authors = []
        Library.library_count += 1  # Increment library count when a new library is created

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, isbn):
        self.books = [book for book in self.books if book.isbn != isbn]

    def list_books(self):
        return [book.title for book in self.books]

    @staticmethod
    def library_info():
        return "Libraries are places that house collections of books."

    @classmethod
    def get_library_count(cls):
        return cls.library_count


# Testing the classes
if __name__ == "__main__":
    # Create authors
    author1 = Author("J.K. Rowling", "1965-07-31")
    author2 = Author("George R.R. Martin", "1948-09-20")

    # Create books
    book1 = Book("Harry Potter and the Sorcerer's Stone", author1.name, "978-0439708180")
    book2 = Book("A Game of Thrones", author2.name, "978-0553103540")

    # Add books to authors
    author1.add_book(book1)
    author2.add_book(book2)

    # Create a library
    library = Library()

    # Add books to the library
    library.add_book(book1)
    library.add_book(book2)

    # List books in the library
    print("Books in the library:", library.list_books())

    # Display book information
    print(book1.display_info("librarian"))
    print(book2.display_info("reader"))

    # Remove a book from the library
    library.remove_book("978-0553103540")
    print("Books in the library after removal:", library.list_books())

    # Get total counts
    print("Total books:", Book.get_total_books())
    print("Total authors:", Author.get_total_authors())
    print("Total libraries:", Library.get_library_count())
```
